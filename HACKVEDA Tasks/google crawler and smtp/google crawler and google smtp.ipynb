```python
import requests
import json
import pandas as pd

questions_list = []

keyword = input("Enter a keyword: ")
keyword1 = keyword


#Multiple site search
site_list = ["quora.com","reddit.com"] # Fixed typo in 'reddit.com'
for i in site_list:
  keyword = keyword+" site:"+i # Added a space before 'site:' for better query formatting
  #key file = your google api key
  #cx = custom search engine id
  google_url = "https://customsearch.googleapis.com/customsearch/v1/?key=your google api key&cx=custom search engine id"
  google_url = google_url + "&q="+keyword

  response = requests.get(google_url)

  # Check for successful API response
  if response.status_code == 200:
    json_response = json.loads(response.text)

    # Check if search results are present
    if 'searchInformation' in json_response:
      total_results = int(json_response ["searchInformation"]["totalResults"])
      #get the start index of next page if it exists
      if 'nextPage' in json_response['queries']:
        next_Index = json_response['queries']['nextPage'][0]['startIndex']
      else:
        next_Index = None # Handle the case when there is no next page

      #Loop to all the responses from google
      total_pages = round(total_results/10)
      try:
        #Loop through the responses
        for item in json_response["items"]:
            title = item["title"]
            title = title.replace(" - Quora","")
            questions_list.append(title)

      except Exception as e:
        pass
    else:
      print(f"Error: No search results found. Check your API key, cx, and query. Response: {json_response}")
  else:
    print(f"Error fetching results from Google API. Status code: {response.status_code}")

  #reset the keyword
  keyword = keyword1

questions_dict = {"Questions": questions_list}
df = pd.DataFrame(data = questions_dict)
df
```
